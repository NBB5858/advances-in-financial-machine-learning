# Exercises

### 11.1
An analyst fits an RF classifier where some of the features include seasonally
adjusted employment data. He aligns with January data the seasonally adjusted
value of January, etc. What “sin” has he committed?

The sin is lookahead bias. Employment data is typically revised after its initial release, as more firms respond to the survey sent out by the Bureau of Labor Statistics. Also, firms that are more competent and better run will generally respond to the surveys first.  Also, only a sample of firms are ever surveyed; the total jobs estimate comes from taking employment data from the survey and scaling it up to represent the total number of firms. When then initial jobs numbers are released, the right scale factor is not known.  If the economy is contracting, it's likely that firms are going out of business and the scale factor is too high. Vice versa if the economy is growing.

### 11.2 
An analyst develops an ML algorithm where he generates a signal using closing
prices, and executed at close. What’s the sin?

The sin is look-ahead bias; the model is trained on information that is not available (or is just barely available) when an action in the strategy needs to be executed

### 11.3
There is a 98.51% correlation between total revenue generated by arcades and
computer science doctorates awarded in the United States. As the number of
doctorates is expected to grow, should we invest in arcades companies? If not,
what’s the sin?

Correlation does not imply causation, and this is likely a spurious correlation detected through multiple testing that was not properly accounted for.

### 11.4
The Wall Street Journal has reported that September is the only month of the
year that has negative average stock returns, looking back 20, 50, and 100 years.
Should we sell stocks at the end of August? If not, what’s the sin?

The sin is potentially cherry picking look back periods over which to calculate the average stock return. We have no way of knowing what other look back periods this fails for. When these results are reported, the total number of hypotheses shoudl also be stated as well.

### 11.5
We download P/E ratios from Bloomberg, rank stocks every month, sell the top
quartile, and buy the long quartile. Performance is amazing. What’s the sin?

The sin is survivorship bias. The current universe of stocks may not be the actual universe of stocks present during the backtest period. Specifically, ignoring survivorship biases towards focusing on companies that are more succesful and longer lasting than average
